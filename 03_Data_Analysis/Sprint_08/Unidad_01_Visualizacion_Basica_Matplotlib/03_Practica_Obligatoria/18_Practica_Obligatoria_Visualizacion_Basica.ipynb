```python
df_california=pd.read_csv("./data/california_cities.csv")
df=df_california.copy()
#Mostramos contenido y descripcin
df.head(10)
df.info()
df.describe()
```

```python
#Descripcin de las columnas numricas
df.describe()
```

```python
#Consultamos los nulos por campo
df.isna().sum().sort_values(ascending=False)
```

```python
#Sustituimos nico valor nulo de la columna area_water_sq_mi
df.loc[df.area_water_sq_mi.isna(),"area_water_sq_mi"]=df.loc[df.area_water_sq_mi.isna(),"area_total_sq_mi"]-df.loc[df.area_water_sq_mi.isna(),"area_land_sq_mi"]
```

```python
df.loc[df.area_total_sq_mi.isna(),"area_total_sq_mi"]=df.loc[df.area_total_sq_mi.isna(),"area_land_sq_mi"]+df.loc[df.area_total_sq_mi.isna(),"area_water_sq_mi"]
```

```python
#Sustituimos el nulo de area_total_km2 para la ciudad 163 (Gridley)
df.loc[163,"area_total_km2"]=df.loc[163, "area_water_km2"]+df.loc[163, "area_land_km2"]
```

```python
#Sustituimos 2 nulos de area_water percent
df.loc[df.area_water_percent.isna(),"area_water_percent"]=df.loc[df.area_water_percent.isna(),"area_water_km2"]/df.loc[df.area_water_percent.isna(),"area_total_km2"]*100
```

```python
df["relacion_km2_sq_mi"]=df.area_total_km2/df.area_total_sq_mi
relacion1=df.relacion_km2_sq_mi.mean()
relacion1
```

```python
#Hay 4 ciudades que tienen nulos en: area total, land y water (km2).
condicion= df.area_total_km2.isna()
df.loc[condicion,["area_total_km2"]]=df.loc[condicion, "area_total_sq_mi"]*relacion1
df.loc[condicion,["area_land_km2"]]=df.loc[condicion, "area_land_sq_mi"]*relacion1
df.loc[condicion,["area_water_km2"]]=df.loc[condicion, "area_water_sq_mi"]*relacion1
```

```python
#Sustituimos los nulos de area water pecent.
df.loc[df.area_water_percent.isna(),"area_water_percent"]=df.loc[df.area_water_percent.isna(),"area_water_km2"]/df.loc[df.area_water_percent.isna(),"area_total_km2"]*100
```

```python
df["relacion_ft_m"]=df.elevation_ft/df.elevation_m
```

```python
relacion2=df.relacion_ft_m.mean()
relacion2
```

```python
#Sustituimos los nulos de elevation_m
condicion2=(df.elevation_m.isna() & df.elevation_ft.notna())
df.loc[condicion2,"elevation_m"]=df.loc[condicion2,"elevation_ft"]/relacion2
```

```python
#Sustituimos los nulos de elevation_ft
condicion3=(df.elevation_m.notna() & df.elevation_ft.isna())
df.loc[condicion3,"elevation_ft"]=df.loc[condicion3,"elevation_m"]*relacion2
```

```python
condicion4=(df.elevation_m.isna() & df.elevation_ft.isna())
df.loc[condicion4, :]
```

```python
#Creamos diccionario de altitudes que faltan
elevaciones_ciudades = {
    "CarmelbytheSea": 77,
    "Claremont": 356,
    "JurupaValley": 236,
    "LaCaadaFlintridge": 362,
    "Moraga" : 152,
    "SanMarcos" :177,
    "UnionCity": 21,
    "Ventura": 11,
    # agrega aqu los dems cuando los tengas
}
```

```python
# Condicin: la ciudad est en el diccionario y elevation_m es NaN
mask = df["city"].isin(elevaciones_ciudades.keys()) & df["elevation_m"].isna()
df.loc[mask, "elevation_m"] = df.loc[mask, "city"].map(elevaciones_ciudades)
```

Recrea la siguiente figura, para ello construye en apartados separados cada grfico y luego haz un apartado en el que se junten los cuatro.  

Las figuras son:

```python
ciudades=df[["city","population_total"]].sort_values(by="population_total",ascending=False).head(5)
```

```python
area=df[["city","area_total_km2"]]
```

```python
area_agua=df[["area_water_km2","population_total"]]
```

```python
area_agua.loc[(area_agua.area_water_km2 > 80)]
```

```python
altitud=df.elevation_m
```

```python
#Creamos la figura con subplots:
fig,ax=plt.subplots(2,2,figsize=(20,20))
ax[0,0].set_title("Ciudades ms pobladas de California", fontsize=20)
bars=ax[0,0].bar(ciudades.city,ciudades.population_total, color=["steelblue","lightgrey", "salmon","olive", "brown"], alpha=0.8)
ax[0,0].set_ylabel("Habitantes (millones)", fontsize=18)
ax[0,0].bar_label(bars)
ax[0,1].set_title("Area total ocupada", fontsize=20)
ax[0,1].hist(area.area_total_km2, bins=50, color="steelblue")
ax[0,1].set_xlabel("superficie(km$^2$)",fontfamily="Bradley Hand ITC", fontsize=18)
ax[0,1].set_ylabel("Nmero de ciudades",fontfamily="Bradley Hand ITC", fontsize=18);
ax[0,1].legend()
ax[1,0].set_title("Poblacin por km$^2$ de agua", fontsize=20)
ax[1,0].scatter(area_agua.area_water_km2,area_agua.population_total, color="green")
ax[1,0].set_xlabel("Area de agua en la ciudad (km$^2$)", fontsize=18)
ax[1,0].set_ylabel("Poblacin (millones de habitantes)", fontsize=18)
ax[1,0].set_xlim(right=130)
ax[1,0].text(50, 3e6, "Outliers", fontsize=18)
# Flecha 1
ax[1,0].annotate("", xy=(88.0, 3884307), xytext=(55, 3.2e6),arrowprops=dict(arrowstyle="->", color="black", linewidth=1))

# Flecha 2
ax[1,0].annotate("", xy=(122.27, 1345895), xytext=(55, 2.9e6), arrowprops=dict(arrowstyle="->", color="black", linewidth=1))

ax[1,1].set_title("Diagrama de caja de la altitud de las ciudades de California", fontsize=16)
ax[1,1].boxplot(altitud, vert=False, patch_artist=True,  boxprops=dict(facecolor="red", color="black"), medianprops=dict(color="black"))
ax[1,1].set_xlabel("Altitud (m)", fontsize=18);
plt.savefig("fig.png", dpi=300, bbox_inches="tight")

```
