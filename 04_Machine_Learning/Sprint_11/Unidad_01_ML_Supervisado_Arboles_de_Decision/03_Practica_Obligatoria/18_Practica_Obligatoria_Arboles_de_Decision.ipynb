```python
#Aplicamos el test de chi2 para determinar la relacin de las columnas categricas con la variable target
from scipy.stats import chi2_contingency

def test_chi2(df, target, cat_cols):
    results = []
    for col in cat_cols:
        contingency_table = pd.crosstab(df[col], df[target])
        chi2, p, dof, ex = chi2_contingency(contingency_table)
        results.append({'Variable': col, 'p-value': p})
    
    return pd.DataFrame(results).sort_values(by='p-value')

# Uso:
print(test_chi2(df, 'y', col_cat))
```

```python
for col in ['age', 'balance', 'duration', 'campaign']:
    plot_categorical_numerical_relationship(df, 'y', col, measure='mean', show_values=True)
```

```python
#Aplicamos transformaciones:
#Transformamos el target a numrico:
df['y'] = df['y'].map({'yes': 1, 'no': 0})

#Aplicamos One-Hot Encoding a las categricas:
df_final = pd.get_dummies(df, dtype=int, drop_first=True)

#Dividimos el dataset en X(features) e y(target).
X = df_final.drop(columns=['y'])
y = df_final['y']

#Split: 80% entrenamiento, 20% test
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test= train_test_split(X, y, random_state=42, stratify=y) #Stratify mantiene la proporcin yes/no

print(f"Set de entrenamiento: {X_train.shape}")
print(f"Set de test: {X_test.shape}")
```

```python
#Importamos la funcin de clasificacin:
from sklearn.tree import DecisionTreeClassifier

#Instanciamos el modelo sin hiperparmetros:
tree_clf=DecisionTreeClassifier(max_depth=5,random_state=42)

#y lo entrenamos:
tree_clf.fit(X_train, y_train)
```

```python
#Importamos la funcin classification_report:
from sklearn.metrics import classification_report

# A. Realizamos las predicciones sobre el juego de TEST (el examen)
y_pred = tree_clf.predict(X_test)

#Generamos el informe comparando las respuestas reales (y_test) con las que dio el modelo (y_pred)
report=classification_report(y_test, y_pred)
print(report)
```

```python
#MODELO BASELINE: Regresin logstica
from sklearn.linear_model import LogisticRegression

#Instanciamos el modelo y los entrenamos
log_reg = LogisticRegression(max_iter=5000, random_state=42)
log_reg.fit(X_train, y_train)

#Lo Evaluamo:
y_pred_log = log_reg.predict(X_test)

print("---Informe de regresin logstico ---")
print(classification_report(y_test, y_pred_log))
```

```python
#Optimizamos los hiperparmetros del rbol:
from sklearn.model_selection import GridSearchCV

#Definimos el espacio de bsqueda
param_grid = {'max_depth': [3, 5, 7, 10, None],'min_samples_split': [2, 10, 20],'criterion': ['gini', 'entropy']}

#Configuramos la bsqueda
grid_tree = GridSearchCV(
    DecisionTreeClassifier(random_state=42),
    param_grid,
    cv=5, 
    scoring='f1' # Usamos F1 porque nos importa detectar los "yes"
)

#Entrenamos la bsqueda
grid_tree.fit(X_train, y_train)

print(f"Mejores parmetros: {grid_tree.best_params_}")
```

```python
#Dibujamos el rbol optimizado:
from sklearn.tree import plot_tree

best_tree = grid_tree.best_estimator_
plt.figure(figsize=(20,10))
plot_tree(best_tree);
```

```python
#Importamos el subpaquete:
from sklearn.metrics import f1_score, accuracy_score

#Predicciones finales del rbol optimizado
y_pred_best = best_tree.predict(X_test)

#Comparativa rpida
print(f"F1-Score Baseline (LogReg): {f1_score(y_test, y_pred_log):.3f}")
print(f"F1-Score rbol Inicial:     {f1_score(y_test, y_pred):.3f}")
print(f"F1-Score rbol Optimizado:  {f1_score(y_test, y_pred_best):.3f}")
```
